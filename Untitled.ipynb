# pivot table

```python
import pandas as pd
import numpy as np
from pydataset import data
mtcars= data('mtcars')
mtcars.shape
```

```python
mtcars.head()
```

```python
mtcars.pivot_table(values = 'mpg', index='gear', observed = False)
# Mean mpg of Gear 3 cars is 16
```

```python
mtcars.pivot_table(values = 'hp', index='cyl', observed = False)# mean hp of Cyl-4 cars is 82
```

```python
mtcars.pivot_table(values= ['mpg','hp'],index = ['gear','am'],observed= False,
                   aggfunc=['sum','mean','count']).reset_index()
#sum of mpg for each gear type
```

```python
mtcars.pivot_table
```

```python
pd.pivot_table(data=mtcars, observed=False,
              index='gear',     #rows
              values='mpg',     #Values to summarize             
              aggfunc='median')  #Aggregation function
# Median of Gear-3 cars is 15.5
```

```python
pivot1= pd.pivot_table(
    data=mtcars, observed= False,
    index='gear',
    columns='am',
    values='mpg',
    aggfunc='mean'
)
```

```python
pivot1 # Mean of Gear-3 & Auto Tx - 16, while for Gear-3 & Manual Tx- No cars
```

```python
print(mtcars.am.value_counts(), mtcars.cyl.value_counts())
```

```python
pd.pivot_table(observed= True,  data=mtcars,  index='gear',  columns='am', \
values='mpg',
               aggfunc=['mean','max','min','count','sum']\
              )
# 32 cars
```

```python
pd.pivot_table( data=mtcars,  index=['gear','am'],  columns='cyl', 
values='mpg',\
               aggfunc=['mean','max','min','count','sum','std']).reset_index()
```

```python
pd.pivot_table(data=mtcars, index='cyl', columns='carb',values='hp', aggfunc=['min','max','count'])
```
