```python
df
```

# wide to long

```python
df_long = df.melt(id_vars=['Rollno','Name'], 
                 value_vars=['Maths','English'], 
                 var_name='Subject', 
                 value_name='Marks')
df_long
```

```python
wide_df = df_long.pivot(index=['Rollno','Name'], columns='Subject',values='Marks').reset_index(drop= False).rename_axis(None, axis=1)
wide_df
```

```python
wide_df2 = df_long.pivot_table(index =['Rollno','Name'],columns='Subject',values='Marks',
                               aggfunc='sum').reset_index(drop= False).rename_axis(None,axis=1)
wide_df2
```

```python
df_long.groupby('Subject') ['Marks'].agg(['sum','mean','count'])
```

# Missing Values

```python
import numpy as np
data = [
    [1, 'Taheera', None, 86],
    [2, 'Aelita', 56, 90],
    [3, 'Khulan', np.nan, None],
    [4, 'Batgerel', 71, 79],
    [5, 'Manisha', np.nan, 76],
]

df = pd.DataFrame(data, columns=['Rollno','Name', 'Maths', 'English'])
df
```

```python
df.isna().sum()
```

```python
df.isna().sum(axis=1) 
```

```python
df.fillna(40)
```

```python
df['Maths']. mean()
```

```python
df['Maths']= df['Maths']. fillna(df['Maths'].mean())
df
```

```python
df['English']= df['English']. fillna(df['English'].min())
df
```

```python
df.loc[df['Rollno']== 1,'Maths'] = 88
df.loc[df['Rollno']== 3,['Maths','English']] = [100,100]
df
```
