```python
data = {
    "car_name": [ "Mazda RX4", "Mazda RX4    Wag", "Datsun 710","Hornet 4 Drive", "Hornet   Sportabout  ", "Valiant", "Duster 360", "merc 240D", "Merc 230", "Merc 280" ],
    "mpg": [21.0, 21.0, 22.8, 21.4, 18.7, 18.1, 14.3, 24.4, 22.8, 19.2]}
df= pd.DataFrame(data)
print('Orginal Data:\n',df)
```

```python
mazda_cars=df[df['car_name'].str.contains('Mazda')]
print("\nCars with 'Mazda':\n", mazda_cars)
```

```python
df[df['car_name'].str.contains("Merc", case= True)]
```

```python
df[df['car_name'].str.contains("merc", case=False)]
```

```python
merc_cars = df[df['car_name'].str.startswith("Merc")]
print("\nCars starting with 'Merc':\n", merc_cars)
```

```python
df[df['car_name'].str.startswith(('Merc','merc'))]
```

```python
df[df['car_name'].str.endswith('Drive')]
```
