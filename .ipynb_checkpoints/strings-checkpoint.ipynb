```python
# 1. Creating Strings
# -------------------------------
print("1. Creating Strings")
s1 = "Hello"  #double quote
s2 = 'World'  #single quote
s3 = """This is a multiline 
       String"""  #triple quote (single or double)
print(s1)
print(s2)
print(s3)
```

```python
ms = '''In Python, a string is a sequence of characters used to represent text 
GeeksforGeeks
GeeksforGeeks
 +1
. Strings are one of the most fundamental and frequently used data types in the language'''
print(ms)
```

```python
# 2. String Concatenation
# -------------------------------
print("\n2. Concatenation")
full = s1 + " " + s2
print("Concatenated:", full)
```

```python
sc ='Aelita' + '....' + "Khulan"
print(sc)
```

```python
# 3. String Repetition
# -------------------------------
print("\t\t3. Repetition")
repeat = "Hi .  How are you! " * 5
print(repeat, end=' : I have reached end of string')
```

```python
# 4. Indexing (Access characters)
# -------------------------------
print('\n')
print("4. Indexing")
text = "PythonA"
print(' The string is ', text)
print("First character:", text[0])
print("Last character:", text[-1])
print("3rd character:", text[2])
```

```python
round(len(text)/2)
```

```python
print('Actual String ', text, ' Reverse String is ', text[::-1]) 
```

```python
print("Middle character:", text[round(len(text)/2)-1])
```

```python
# 5. Slicing (Get part of string)
# string[start:stop:step]
# -------------------------------
print("\n5. Slicing")
print("First 3 chars:", text[0:3])
print("From index 2:", text[2:])
print("Reverse string:", text[::-1])
```

```python
#  6. Strings
# -------------------------------
print("\n String Methods")
name = "  dAta science 1syed hasan "
print("Original String with spaces - ", name)
print("Upper:", name.upper())
print("Lower:", name.lower())
print("Strip spaces:", name.strip())
#print("Sentence case :" , name.capitalize())
print("title case :" , name.title())
print("Replace:", name.replace("science", "analysis"))
```

```python
# 7. Checking Content
# -------------------------------
print("\n7. Checking Content")
sample = "Python123"
print("Is alphabet:", sample.isalpha())
print("Is digit:", sample.isdigit())
print("Is alphanumeric:", sample.isalnum())
```

```python
# 8. Splitting and Joining
# -------------------------------
print("\n8. Split & Join")
sentence = "Python is easy to learn"
print(' the Sentence is ', sentence)
words = sentence.split(" ")
print("Split:", words)
joined = "*".join(words)
print("Joined:", joined)
```

```python
# 9. String Formatting
# -------------------------------
print("\n9. Formatting")
name = "Taheera"
age = 25
gender ='F'
city ='New York'
# Old style
print("My name is %s and I am %d years old, and my gender is %s" % (name, age,gender))
# format()
print("My name is {} and I am {} years old, with gender is {} and city in {}".format(name, age, gender, city))
# f-string (recommended)
print(f"My gender is {gender}, name is {name} and I am {age} years old")
```

```python
# 11. Important Concept: Strings are Immutable
# -------------------------------
print("\n11. Immutability")
text = "hello"
# text[0] = 'H'  # ❌ This will give an error
# Correct way:
text = "H" + text[1:]
print("Modified string:", text)
print("He said \"Python is fun\"")  #double quotes
```

```python
text[0] = 'H'
```

```python
# 12. Length of String
# -------------------------------
print("\n12. Length")
print("Length of 'Python':", len("Python"))
```

```python
# 13. Membership (in keyword)
# -------------------------------
print("\n13. Membership")
print("Py" in "Python")
print("Java" not in "Python")
```

```python
import pandas as pd
from pydataset import data
```

```python
data = {
    "car_name": [ "Mazda RX4", "Mazda RX4    Wag", "Datsun 710","Hornet 4 Drive", "Hornet   Sportabout  ", "Valiant", "Duster 360", "merc 240D", "Merc 230", "Merc 280" ],
    "mpg": [21.0, 21.0, 22.8, 21.4, 18.7, 18.1, 14.3, 24.4, 22.8, 19.2]}
df = pd.DataFrame(data)
print("Original Data:\n", df)
```

```python
Mazda_cars= df[df['car_name'].str.contains('Mazda')]
print('\ncars with Mazda', Mazda_cars)
```

```python
df[df['car_name'].str.contains("Merc", case= False)]
```

```python
Hornet_cars = df[df['car_name'].str.startswith("Hornet")]
print("\nCars starting with 'Merc':\n", Hornet_cars)
```

```python
df[df['car_name'].str.startswith(("Merc",'merc'))]
```

```python
df[df['car_name'].str.endswith(('Drive'))]
```

```python
# Replace Strings
df['car_name_updated'] = df['car_name'].str.replace("Merc", "Mercedes", regex=False)
print("\nUpdated Names:\n", df[['car_name', 'car_name_updated']])
```

```python
df['car_name'].str.replace("Merc", "Mercedes", case=False, regex=False)
```

```python
df['brand'] = df['car_name'].str.split().str[0]
print("\nExtracted Brand:\n", df[['car_name', 'brand']])
#👉 Very useful for: * extracting categories, * feature engineering
```

```python
df['upper_name'] = df['car_name'].str.upper()
df['lower_name'] = df['car_name'].str.lower()
df
```

```python
# Cleaning (trim spaces)
df['clean_name'] = df['car_name'].str.strip()
df
```
