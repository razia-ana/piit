```python
mtcars = data('mtcars')
mtcars = mtcars.astype({ 'cyl':'category', 'vs': 'category',  'carb': 'category',  
                         'am':   'category',  'gear':  'category' } )
```

```python
mtcars['vs'] = mtcars['vs'].map({0: 'vshaped', 1: 'Straight'})
mtcars.head()
```

```python
mtcars['am'] = mtcars['am'].map({0:'auto',1: 'manual'})
mtcars.head()
```

# pivot table
-index: what goes down the rows
-columns: what goes accross the top
-values: what fills the table 
-aggfunc: how to summarice if there duplicates

```python
mtcars.head()
```

```python
mtcars.pivot_table(values='mpg',index='gear',observed = False)
```

```python
mtcars.pivot_table(values= 'hp', index= 'cyl',observed= False)
```

```python
mtcars.pivot_table(values= ['mpg','hp'],index=['cyl','am'], observed= True,
                    aggfunc=['sum','mean','count']).reset_index()
```

```python
pd.pivot_table(data= mtcars,observed= False,
               index= 'gear',
               values='mpg',
               aggfunc= 'median')
```

```python
piovt1= pd.pivot_table(data=mtcars,observed= False,
                      index='gear',
                      columns='am',
                      values='mpg',
                      aggfunc='mean')
```

```python
piovt1
```

```python
print(mtcars.am.value_counts(),mtcars.cyl.value_counts())
```

```python
pd.pivot_table( observed=True , data= mtcars,  index='gear',  columns='am', \
                values='mpg',
                aggfunc=['mean', 'max', 'min' ,'count','sum']\
              )
```

```python
pd.pivot_table(data=mtcars,index='cyl',columns='carb', values='hp',aggfunc=['min','max','count'],observed=False)
```

```python
mtcars.pivot_table(
    index= 'gear',
    columns='am',
    values=['hp','mpg'],
    aggfunc='mean')
```

```python
mtcars.pivot_table(
    index= 'cyl',
    columns='gear',
    values='hp',
    aggfunc=['min','max']
)
```

```python
mtcars.pivot_table(
    index= 'cyl',
    columns='carb',
    aggfunc='size',
    fill_value= 0,)
```
