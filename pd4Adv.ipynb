```python
bins = [0, 17, 29, mtcars['mpg'].max()]
labels = ['Low', 'Medium', 'High']
# Create the new categorical column
mtcars['mpg_level'] = pd.cut(mtcars['mpg'], bins=bins, labels=labels, include_lowest=True)
```

```python
mtcars.describe ()
```

```python
bins = [0, 70, 150, 250, mtcars['hp'].max()]
labels = ['Low', 
          'Medium', 'High', 'VeryHigh']
mtcars['hp_level'] = pd.cut(mtcars['hp'], bins=bins, labels=labels, include_lowest=True)
```
